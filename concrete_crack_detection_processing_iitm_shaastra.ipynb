```python
from os import listdir
from os.path import isfile, join
import numpy
import cv2

mypath='/content/processing_material'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = numpy.empty(len(onlyfiles), dtype=object)
for i in range(0, len(onlyfiles)):
  images[i] = cv2.imread( join(mypath,onlyfiles[i]) )
  # plt.imshow(images[i])
  # plt.show()
```

```python
#bilateral filter is working well 
for i in range(0, len(onlyfiles)):
  plt.imshow(images[i])
  plt.show()
  images[i] = cv2.bilateralFilter(images[i],6, 75,1475)#75
  plt.imshow(images[i])
  plt.show()
```

```python
for i in range(0, len(onlyfiles)):
  plt.imshow(images[i])
  plt.show()
  kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
  images[i] = cv2.morphologyEx(images[i], cv2.MORPH_OPEN, kernel, iterations=1)
  # images[i] = cv2.morphologyEx(images[i], cv2.MORPH_CLOSE, kernel, iterations=2)
  plt.imshow(images[i])
  plt.show()
```
